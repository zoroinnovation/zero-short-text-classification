```python
from datasets import load_dataset
import re

# Load SST-2 dataset
dataset = load_dataset("glue", "sst2")

# Preview raw example
dataset['train'][0]
```

```python
def format_sst2(example):
    return {
        "text": example["sentence"],
        "label": example["label"]  # 0 = Negative, 1 = Positive
    }

formatted_train = dataset['train'].map(format_sst2)
formatted_val = dataset['validation'].map(format_sst2)
```

```python
def clean_text(text):
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove special chars
    text = re.sub(r"\s+", " ", text).strip()     # Normalize spaces
    return text

formatted_train = formatted_train.map(lambda x: {
    "text": clean_text(x["text"]),
    "label": x["label"]
})

formatted_val = formatted_val.map(lambda x: {
    "text": clean_text(x["text"]),
    "label": x["label"]
})
```

```python
import pandas as pd

# Convert HuggingFace Dataset to Pandas DataFrame
df_train = pd.DataFrame(formatted_train)
df_val = pd.DataFrame(formatted_val)

# Save to CSV
import os

os.makedirs("data_csv", exist_ok=True)
df_train.to_csv(r"D:\Zoro_project2\zero-short-text-classification\data_csv/sst2_train.csv", index=False)
df_val.to_csv(r"D:\Zoro_project2\zero-short-text-classification\data_csv/sst2_val.csv", index=False)

print(" Saved to data_csv/ successfully.")

```
